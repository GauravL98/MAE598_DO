```python
from bayes_opt import BayesianOptimization

def fun(x1,x2):
  return -((4 - 2.1*(x1**2) + (1/3)*(x1**4))*(x1**2) + x1*x2 + ((-4 + 4*(x2**2))*(x2**2)))

bounds = {'x1':(-3,3),'x2':(-2,2)}

optimizer = BayesianOptimization(f=fun,pbounds=bounds,random_state=1)
optimizer.maximize(init_points=25,n_iter=200)

min_value = -optimizer.max['target']
min_x = optimizer.max['params']
print('objective funtion minimum value=',min_value)
print('x1 and x2 at minmum fuction value=',min_x)
```
